```python
# Note, the transpose and dropping duplicates has changed all the data types to 'object'. 
# We use a function here as we need to use this again later on.

def data_type_converter(dataframe):
    dataframe.loc[:,'Precincts':'Injury_deaths'] = df_data.loc[:,'Precincts':].astype(float)
    dataframe['Fips'] = df_data['Fips'].astype(int)
    dataframe['State'] = df_data['State'].astype(str)
    dataframe['County'] = df_data['County'].astype(str)
    dataframe['ST'] = df_data['County'].astype(str)

data_type_converter(df_data)
df_data.info()
```

### Question 4
Finally, augment the data so that each county contains the relative vote count for each party in each election. (e.g. For each county, if the Republicans got 2300 votes and the Democrats got 1900 votes, then the relative vote shares should be 54% and 46%.)

<div style = "background-color: #ddffff" >
Here we joined both tables together and then created 3 new columns. Each column shows the percentage of democrat votes in 2008, 20012 and 2016. The percentage of republican votes is just 1 minus the percentage of democrat votes.
</div>

```python
# Joining the 2 tables together
df_votes_data = df_votes.merge(df_data, on='Fips', how='left')

# Fraction of democrat votes in 2008
df_votes_data['fraction_democrats_08'] = df_votes_data['democrats_08_votes'] / \
(df_votes_data['democrats_08_votes']+df_votes_data['republicans_08_votes'])

# Fraction of democrat votes in 2012
df_votes_data['fraction_democrats_12'] = df_votes_data['democrats_12_votes'] / \
(df_votes_data['democrats_12_votes']+df_votes_data['republicans_12_votes'])

# Fraction of democrat votes in 2016
df_votes_data['fraction_democrats_16'] = df_votes_data['democrats_16_votes'] / \
(df_votes_data['democrats_16_votes']+df_votes_data['republicans_16_votes'])


# Displays the counties with the highest fraction of democrat votes
display(df_votes_data.loc[:,('State',  'County', 'fraction_democrats_08', 'fraction_democrats_12', 'fraction_democrats_16')]\
.sort_values(by=['fraction_democrats_16'],  ascending=False).head(10))

# Displays the counties with the lowest fraction of democrat votes
display(df_votes_data.loc[:,('State',  'County', 'fraction_democrats_08', 'fraction_democrats_12', 'fraction_democrats_16')]\
.sort_values(by=['fraction_democrats_16'],  ascending=False).tail(10))

```

What is the fraction of democrats per state?

<div style = "background-color: #ddffff" >
Bellow is a bar chart showing the fraction of democrats per state in 2016. Bellow this are some more bar charts looking at how pro-republican states vs pro-democrat states differ on education, people uninsured, unemployment and obesity. These graphs (other than the bar chart showing the fraction of democrats per state) are not representing the population but rather comparing the counties.
```python
#Once again we need to reset the data types
data_type_converter(df_votes_data)

# Create a new columns with the votes*fraction_of_democrats
df_votes_data['np_percent_total_16'] = df_votes_data['Votes'] * df_votes_data['fraction_democrats_16']

# New data frame with the sum of all columns grouped by state
df_sum_votes_data = df_votes_data.groupby(['State']).sum()

# Create a new column with the fraction of democrats per State
df_sum_votes_data['fraction_decmocrats_state_16'] = df_sum_votes_data['np_percent_total_16'] / df_sum_votes_data['Votes']
df_sum_votes_data.sort_values(by='fraction_decmocrats_state_16', inplace=True, ascending=False)

# Plots the graph
plt.rcParams['figure.figsize'] = (10,5)
plt.plot()
barchart_y_val = df_sum_votes_data['fraction_decmocrats_state_16']
barchart_x_val = df_sum_votes_data.index
plt.bar(barchart_x_val, barchart_y_val)
plt.xticks( rotation=90)
plt.ylabel('Percentage Democrats', size = 16)
plt.title('Percentage of Democrats per State', size = 25, pad=20)
plt.show()
```

```python
#High school averages over all the different counties.
hight_school_democrats = df_votes_data.loc[df_votes_data['fraction_democrats_16'] < 0.5 ,'At_Least_High_School_Diploma'].mean()
hight_school_republicans = df_votes_data.loc[df_votes_data['fraction_democrats_16'] > 0.5 ,'At_Least_High_School_Diploma'].mean()
print('hight_school_democrats:',hight_school_democrats)
print('hight_school_republicans:',hight_school_republicans)


plt.figure()
plt.bar(['Republicans', 'Democrats'], [hight_school_republicans, hight_school_democrats])
plt.ylim([82, 84])
plt.ylabel('Percent', size = 16)
plt.title('Percentage of High School Diplomas', size = 25, pad=20)
plt.show()
```

```python
# Percentage of people uninsured
Uninsured_democrats = df_votes_data.loc[df_votes_data['fraction_democrats_16'] < 0.5 ,'Uninsured'].mean()*100
Uninsured_republicans = df_votes_data.loc[df_votes_data['fraction_democrats_16'] > 0.5 ,'Uninsured'].mean()*100
print('Uninsured_democrats:',Uninsured_democrats)
print('Uninsured_republicans:',Uninsured_republicans)

plt.figure()
plt.bar(['Republicans', 'Democrats'], [Uninsured_republicans, Uninsured_democrats])
plt.ylabel('Percent', size = 16)
plt.title('Percentage of population Uninsured', size = 25, pad=20)
plt.ylim([16, 19])
plt.show()
```
